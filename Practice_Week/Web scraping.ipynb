```python
import pandas as pd
```

```python
u_cols = ['user_id', 'age', 'sex', 'occupation', 'zip_code']
users = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.user', sep='|', names = u_cols)
users.head()
```

```python
selected_males = users[(users.age==40) & (users.sex=='M')].head()
selected_males.head()
```

```python
#lambda function
grouped_data_lambda = users['sex'].groupby(users.occupation)
male_dominant_occupations = grouped_data_lambda.apply(lambda f: sum(f=='M') > sum(f=='F'))
print(male_dominant_occupations)
```

```python
selected_users = users[(users.sex=='F') & (users.occupation=='programmer')]
selected_users.head()
selected_users.describe()
```

```python
r_cols = ['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings = pd.read_csv('http://files.grouplens.org/datasets/movielens/ml-100k/u.data', sep='\t', names = r_cols)
ratings.head()
```

```python
grouped_data = ratings.groupby('user_id') #split
ratings_per_user = grouped_data.count()
ratings_per_user.head()
```
